Import the necessary libraries such as Pandas

```python
import pandas as pd
```

Retrieve the data from the URL and create a dataframe with the information we only need.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df = pd.read_html(url)
df = df[0]
```

Remove every row where the value in the column "Borough" is "Not assigned"

```python
df = df[df["Borough"] != "Not assigned"]
df.head()
```

The frame already has multiple neighbourhoods in the same row if their postal code is identical

```python
df["Postal Code"]
```

The following code shows there are no rows in this frame which have a neighbourhood with the value "Not assigned". Thus there are no required changes to make in the preprocessing phase of the data

```python
df_notAssignedHoods = df[df["Neighbourhood"] == "Not assigned"]
df_notAssignedHoods.shape[0]
```

Reset the index so the numbers match to the current length of the dataframe

```python
df2 = df.reset_index()
# remove the column with the old values
df2 = df2.drop(["index"], axis = 1)
# check if the code worked as intended
df2.head()
```
